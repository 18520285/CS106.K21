```python
import bs4
import pandas as pd
import requests
import os
```

```python
#Đọc lấy file chứa các đường dẫn tới sản phẩm lấy bình luận
f = open('url_data.txt','r')
data = f.read().split('\n')
f.close()
```

```python
#Kiểm tra 10 đường dẫn dầu
for i in range(10):
    print(data[i])
```

```python
#Tải file html của trang sản phẩm cần lấy comment
def get_page_content(url):
    hdr = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
         'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
         'Referer': 'https://cssspritegenerator.com',
         'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
         'Accept-Encoding': 'none',
         'Accept-Language': 'en-US,en;q=0.8',
         'Connection': 'keep-alive'}
    page = requests.get(url,headers=hdr)
    return bs4.BeautifulSoup(page.text, "html.parser")
```

```python
#Lọc lấy comment từ trang html
def get_comment(url1):
    sl = 0
    soup = get_page_content(url1)
    sp = soup.find_all('li', class_='par')
    for item in sp:
        comment = item.find_all('i')
        a = [i.text for i in comment]
        while ('' in a):
            a.remove('')
        s = a[0].split('\n')
        a[0] = ''
        for i in s:
            a[0] += (' ' + i)
        good = item.find_all('i', class_='iconcom-txtstar')
        if (a[0] not in X):
            X.append(a[0]) # thêm câu nhận xét vào input của data
            # kiểm tra số sao để xác định output của data 1,2,3 sao thì tiêu cực-0 . 4,5 sao thì tích cực-2
            if len(good) in [1, 2, 3]:
                Y.append(0)
            else:
                Y.append(1)
            sl+=1
    print("Get",sl,"comments!")
```

```python
#Lấy dữ liệu
X = [] #Set input
Y = [] #Set output
for index, i in enumerate(data):
    print("Processing code of produce",index+1,"/",len(data),"produce.")
    soup = get_page_content(i)
    page = soup.find_all('div', class_='pagcomment')
    if len(page)>0:
        num_page = [i.text for i in page[0].find_all('a')]
        num_page.remove('»')
        num_page = [int(i) for i in num_page]
        for num in range(1,max(num_page)):
            get_comment(i+'?p='+str(num))
    else:
        get_comment(i)
    print("-"*100)
```

```python
print("Total: crawled",len(X),"comments!")

#Lưu dữ liệu vào CSV file
d = {'comment': X, 'sentiment': Y}
df = pd.DataFrame(data=d)
df.to_csv('comment_data.csv',index=False, sep=',', encoding='utf-8')
```
